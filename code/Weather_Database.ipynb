```python
#Strategy to get 2000 random citi information
# Steps 1-8
# 1.1 in a while loop_cnt < 2000 loop_cnt initialized to 0  
# 1.2 Generate random, latituede longitude pair  using num py; using unifrom distribution within range
#    min/max range lat - -90  to + 90
#    min/max range lng - -180 to + 180
# 1.3 Use citipy library get city, country information
# 1.4 check if city_country pair is unique - use python list.feature - item not in list[]
# 2. Create a new Jupyter Notebook file to retrieve the weather data, and name it Weather_Database.ipynb.
# 3. Create a new set of 2,000 random latitudes and longitudes.
# 4. Get the nearest city using the citipy module.	
# 5. Perform an API call with the OpenWeatherMap.	
# 6. Retrieve the following information from the API call:	
#         % Humidity
#         % Cloudiness
#         Wind Speed
#         Weather Description

# DEBUG STEPS TO DEVELOP THIS CODE
# url = "http://api.openweathermap.org/data/2.5/weather?"
# city="San Ramon"
# units="Imperial"
# city_query_url = f"{url}appid={weather_api_key}&q={city}&units={units}"

# city_weather_resp = requests.get(city_query_url).json()
# # print(json.dumps(city_weather_resp, indent=4, sort_keys=True))

# print(f'temperature {city_weather_resp["main"]["temp"]}')
# print(f'max temp {city_weather_resp["main"]["temp_max"]}')
# print(f'humidity {city_weather_resp["main"]["humidity"]}')
# print(f'Cloudiness {city_weather_resp["clouds"]["all"]}')
# print(f'Wind Speed {city_weather_resp["wind"]["speed"]}')
# print(f'Current description {city_weather_resp["weather"][0]["description"]}')

url = "http://api.openweathermap.org/data/2.5/weather?"
units="Imperial"

n=1
lat_min=-90. 
lat_max=90. 
lng_min=-180.
lng_max=180.
tot_city = 0

city_country_name_list = []
city_name_list = []
city_country_list = []
city_lat_list = []
city_lng_list = []
city_max_temperature_list = []
city_humidity_list = []
city_cloudiness_list = []
city_wind_speed_list = []
city_cur_descript_list = []

# Debug with short number - Establish flow then take a long number 
# 1    DONE - GOOD
# 10   DONE - GOOD
# 200  IP
# 2000
lat         = np.random.uniform(lat_min, lat_max, 5000) # Draw 5000 samples
lng          = np.random.uniform(lng_min, lng_max, 5000) # Draw 5000 samples

lat_lng_idx = 0
inc=0
rev = 0
inc_lat=1
inc_lng=1
while ( tot_city < 2000 and lat_lng_idx < len(lat) and rev < 3): 
    if ( rev ==  0 ): 
        city_n       = citipy.nearest_city(lat[len(lat)-1-lat_lng_idx], lng[lat_lng_idx])
        if ( lat_lng_idx >= (len(lat)-1) ): 
            lat_lng_idx = 0
            rev+=1
            print(f'rev:0 switch rev:{rev} lat_lng_idx{lat_lng_idx}')
    elif ( rev == 1 ): 
        city_n       = citipy.nearest_city(lat[lat_lng_idx], lng[len(lat)-1-lat_lng_idx])
        if ( lat_lng_idx >= (len(lat)-1) ):
            lat_lng_idx = 0
            rev+=1
            print(f'rev:1 switch rev:{rev} lat_lng_idx{lat_lng_idx}')
    elif(  rev == 2 ):
        city_n       = citipy.nearest_city(lat[lat_lng_idx], lng[lat_lng_idx])
        if ( lat_lng_idx >= (len(lat)-1) ): # rev >=2
            lat_lng_idx = 0
            rev+=1
            print(f'rev:2 switch rev:{rev} lat_lng_idx{lat_lng_idx}')
        
    city_name    = city_n.city_name
    country_name = city_n.country_code
    city_query_url = f"{url}appid={weather_api_key}&q={city_name}&units={units}"
    try:
        city_weather_resp = requests.get(city_query_url).json()
    except:
        print(f'Skipping for city:{city_name}')
    ##weather_response = requests.get(query_url)
    ##weather_json = weather_response.json()
    #print(json.dumps(city_weather_resp, indent=4, sort_keys=True))

    # concatenate city country, lat, lng, Max Temp, Humidity, Cloudiness, Wind Speed Current Decription to list
    city_country_name = {city_name,country_name}
    #print(city_country_name)            
    try:
        if ( (city_country_name not in city_country_name_list) ):
            city_max_temperature_list.append(city_weather_resp["main"]["temp_max"])
            city_humidity_list.append(city_weather_resp["main"]["humidity"])
            city_cloudiness_list.append(city_weather_resp["clouds"]["all"])
            city_wind_speed_list.append(city_weather_resp["wind"]["speed"])
            city_cur_descript_list.append(city_weather_resp["weather"][0]["description"])
            city_lat_list.append(city_weather_resp["coord"]["lat"])
            city_lng_list.append(city_weather_resp["coord"]["lon"])
            city_country_name_list.append(city_country_name)
            city_name_list.append(city_name)            
            city_country_list.append(country_name)
            tot_city+=1
            #suppress lat_lng_idx increment here
            #lat_lng_idx+=1
            #logic to find cities near bye when we hit a city with random lat/lng co-ordinates BEGIN
            if ( inc < 21 ):                 
                if ( inc < 10 ):
                    inc_lat=1
                    inc_lng=1
                else:
                    inc_lat=1
                    inc_lng=-1
                if ( inc%2 ):
                    lat[lat_lng_idx]+=inc_lat
                else:
                    lng[lat_lng_idx]+=inc_lng
                #increment always
                inc+=1
                if ( inc == 20 ):
                    #allow lat_lng_idx increment here
                    lat_lng_idx+=1
                    #reset everything
                    inc=0
                    inc_lat=1
                    inc_lng=1
            #logic to find cities near bye when we hit a city with random lat/lng co-ordinates END!
                    
            if ( tot_city%50 == 1 or lat_lng_idx%100 == 1 ):
                print(f'Found city:{city_name} country:{country_name} tot_city:{tot_city} lat_lng_idx:{lat_lng_idx} inc:{inc} rev:{rev}')
        else:
            if ( tot_city%50 == 1 or lat_lng_idx%100 == 1):
                print(f'Observed repeat city:{city_name} country:{country_name} tot_city:{tot_city} lat_lng_idx:{lat_lng_idx} inc:{inc} rev:{rev}')
            lat_lng_idx+=1
            #reset everything on duplicate
            inc=0
            inc_lat=1
            inc_lng=1
    except:
        if ( tot_city%50 == 1 or lat_lng_idx%100 == 1):
            print(f'Invalid Response for  city:{city_name} Skipping @ tot_city:{tot_city} lat_lng_idx:{lat_lng_idx} inc:{inc} rev:{rev}')
        if ( tot_city%50 == 1 or lat_lng_idx%100 == 1):
            print(f'Incr tot_city to :{tot_city}') # this is also index
        # order matters because following should not be added if an exception hits!!
        city_max_temperature_list.append("NaN")
        city_humidity_list.append("NaN")
        city_cloudiness_list.append("NaN")
        city_wind_speed_list.append("NaN")
        city_cur_descript_list.append("NaN")
        city_lat_list.append("NaN")
        city_lng_list.append("NaN")
        city_country_name_list.append(city_country_name)
        city_name_list.append(city_name)            
        city_country_list.append(country_name)
        tot_city+=1
        #reset everything on Invalid
        lat_lng_idx+=1
        inc=0
        inc_lat=1
        inc_lng=1
    # order matters because following should not be added if an exception hits!!
    # before exiting while loop add city/Country

# print(city_country_name_list)
print(f'Total tot_city:{tot_city} lat_lng_idx:{lat_lng_idx}')

# create Data frame Step#8 
city_df = pd.DataFrame({
    'City' : city_name_list,
    'Country' : city_country_list,
    'Lat'  : city_lat_list,
    'Lon' : city_lng_list,
    'Max Temp' : city_max_temperature_list,
    '% Humidity' : city_humidity_list,
    '% Cloudiness' : city_cloudiness_list,
    'Wind Speed' :city_wind_speed_list,
    'Current Decription' : city_cur_descript_list
})
city_df
```
